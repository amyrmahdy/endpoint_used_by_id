```python
import pandas as pd
```

```python
df = pd.read_csv("warehouse.csv")
```

```python
df.head()
```

```python
params = {
    "endpoint" : lambda x: ','.join(sorted(pd.Series.unique(x)))
}
targets = ["jalali_date","input"]
mart = df.groupby(targets).agg(params).reset_index()
rename_columns = {'input' : 'id', 'endpoint' : 'endpoint_list'}
mart.rename(columns = rename_columns, inplace = True)
mart.to_csv("mart.csv")
mart
```
